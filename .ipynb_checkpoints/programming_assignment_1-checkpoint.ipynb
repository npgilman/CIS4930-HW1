<h1>Question 1</h1>
<i>Imagine you are interested in the current state of a research topic, “How have VR
(virtual reality)/AR (augmented reality) techniques been used for education?” and
you have collected the basic information from journal articles in this field in the past
5 years. “articleInfo.csv” includes the following information about each article: “Title,
Year, Author Number, Citation, Source, Abstract, and Type” “authorInfo.csv” includes
the following information about each author: “Author Name, Author Affiliation,
Country, and h-index.”
</i>
</br>
</br>
</br>
Merge “articleInfo.csv” and “authorInfo.csv” into one data frame based on “Article
No.” and fill all empty cells with the value of 0.

```python
import pandas as pd
```

```python
article_csv = pd.read_csv("articleInfo.csv").fillna(0)
author_csv = pd.read_csv("authorInfo.csv").fillna(0)
jointDF = pd.merge(
    article_csv,
    author_csv,
    how="inner",
    on="Article No.",
    left_on=None,
    right_on=None,
    left_index=False,
    right_index=False,
    sort=True,
    suffixes=("_x", "_y"),
    copy=True,
    indicator=False,
    validate=None,
)
```

1. Plot the <i>yearly_publication</i> figure, in which the x-axis is the <i>year</i>, the y-axis is
the number of articles published during that <i>year</i>.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
plt.plot(sorted(article_csv.Year.unique()), article_csv.groupby('Year').count(), marker='s', color='b')

plt.title('Yearly Publications')
plt.xlabel('Year')
plt.ylabel('Publications per Year')
plt.show()
```

2. Plot the <i>yearly_citation</i> figure, in which the x-axis is the <i>year</i>, the y-axis is the
total number of citations during that <i>year</i>.

```python
citations = [article_csv.loc[article_csv['Year'] == 2017, 'Citation'].sum(),
             article_csv.loc[article_csv['Year'] == 2018, 'Citation'].sum(),
             article_csv.loc[article_csv['Year'] == 2019, 'Citation'].sum(),
             article_csv.loc[article_csv['Year'] == 2020, 'Citation'].sum(),
             article_csv.loc[article_csv['Year'] == 2021, 'Citation'].sum(),
             article_csv.loc[article_csv['Year'] == 2022, 'Citation'].sum()]

plt.plot(sorted(article_csv.Year.unique()), citations, marker='s', color='b')

plt.title('Yearly Citations')
plt.xlabel('Year')
plt.ylabel('Citations per Year')
plt.show()
```

3. Plot the figure of the number of publications across countries.

```python
import pygal.maps.world
```

```python
# dictionary to convert between country names and the country codes used by the pygal.maps.world module
country_codes = {
    "Taiwan":"tw",    "Korea":"kr",    "South Africa":"za",    "United Kingdom":"gb",
    "USA":"us",    "Cyprus":"cy",    "China":"cn",    "Germany":"de",
    "Mexico":"mx",    "Italy":"it",    "Ukraine":"ua",    "Ireland":"ie",
    "Australia":"au",    "Hong Kong":"hk",    "Chile":"cl",    "Chile ":"cl",
    "Spain":"es",    "Spain ":"es",    "France":"fr",    "Denamrk":"dk",
    "Denmark ":"dk",    "Canada":"ca",    "Czech Republic":"cz",    "Liechtenstein":"li",
    "Palestine":"ps",    "Russia":"ru",    "Slovakia":"sk",    "Malaysia":"my",
    "Pakistan":"pk",    "India":"in",    "Switzerland":"ch",    "0":"0",
    0:"0"
}
```

```python
worldmap_chart = pygal.maps.world.World()
worldmap_chart.title = 'Publications in countries'

country_cnt = {}
df = jointDF.drop_duplicates(subset=['Article No.'])
for c in df['Country']:
    cntry = country_codes[c]
    if (not(cntry in country_cnt)):
        country_cnt[cntry] = 1
    else:
        country_cnt[cntry] += 1
        
worldmap_chart.add("Publications", country_cnt)
```

<h1>Question 3</h1>
<i>For this question, use the same dataset in the above
question.
</i>
</br>
</br>
</br>
Train a Classification model in which: the independent variables
(inputs) are “ASR_Error,” “IntentError,” “Duration,” and “Gender,”; the dependent
variable (output) is “Purchase.” Use the evaluation metrics we introduced in class to
compare the performance of the following four machine learning classification
algorithms: (1) Logistic Regression, (2) SVM, (3) Naive Bayes, and (4) Random
Forest.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
del df[df.columns[-1]]
```

```python
# dependent
y = df['Purchase'].to_numpy()

# independent vars are ASR_Error, Intent_Error, Gender and Duration
X = df.drop('Purchase', axis = 1).drop('SUS', axis=1).to_numpy()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
scale = StandardScaler()
scaled_X = scale.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(scaled_X, y, test_size = 0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix, roc_curve, auc, classification_report

import matplotlib.pyplot as plt
```

```python
lc = LogisticRegression()
svc = SVC(probability=True)
nbc = GaussianNB()
rfc = RandomForestClassifier()
```

```python
lc.fit(X_train, y_train)
svc.fit(X_train, y_train)
nbc.fit(X_train, y_train)
rfc.fit(X_train, y_train)
```

```python
y_lc_predicted = lc.predict(X_test)
y_lc_pred_proba = lc.predict_proba(X_test)

y_svc_predicted = svc.predict(X_test)
y_svc_pred_proba = svc.predict_proba(X_test)

y_nbc_predicted = nbc.predict(X_test)
y_nbc_pred_proba = nbc.predict_proba(X_test)

y_rfc_predicted = rfc.predict(X_test)
y_rfc_pred_proba = rfc.predict_proba(X_test)
```

```python
print(classification_report(y_test, y_lc_predicted))
print(classification_report(y_test, y_svc_predicted))
print(classification_report(y_test, y_nbc_predicted))
print(classification_report(y_test, y_rfc_predicted))

models = ['Logistic Regression','Support Vector Machine', 'Naive Bayes Classifier', 'Random Forest Classifier']
predictions = [y_lc_predicted, y_svc_predicted, y_nbc_predicted, y_rfc_predicted]
pred_probabilities = [y_lc_pred_proba, y_svc_pred_proba, y_nbc_pred_proba, y_rfc_pred_proba]

plot = 1

for model, prediction, pred_proba in zip(models, predictions, pred_probabilities):
    disp = ConfusionMatrixDisplay(confusion_matrix(y_test.ravel(), prediction))
    disp.plot(
        include_values=True,
        cmap='gray',
        colorbar=False
    )
    disp.ax_.set_title(f"{model} Confusion Matrix")

plt.figure(figsize=(30, 15))
plt.suptitle("ROC Curves")
plot_index = 1

for model, prediction, pred_proba in zip(models, predictions, pred_probabilities):
    fpr, tpr, thresholds = roc_curve(y_test, pred_proba[:, 1])
    auc_score = auc(fpr, tpr)
    plt.subplot(3, 2, plot_index)
    plt.plot(fpr, tpr, 'r', label='ROC curve')
    # pyplot.figure(figsize=(5, 5))
    plt.title(f'Roc Curve - {model} - [AUC - {auc_score}]', fontsize=14)
    plt.xlabel('FPR', fontsize=12)
    plt.ylabel('TPR', fontsize=12)
    plt.legend()
    plot_index += 1
plt.show()
```

```python
y_lc_predicted = lc.predict(X_test)
y_lc_pred_proba = lc.predict_proba(X_test)

y_svc_predicted = svc.predict(X_test)
y_svc_pred_proba = svc.predict_proba(X_test)

y_nbc_predicted = nbc.predict(X_test)
y_nbc_pred_proba = nbc.predict_proba(X_test)

y_rfc_predicted = rfc.predict(X_test)
y_rfc_pred_proba = rfc.predict_proba(X_test)

print(classification_report(y_test, y_lc_predicted))
print(classification_report(y_test, y_svc_predicted))
print(classification_report(y_test, y_nbc_predicted))
print(classification_report(y_test, y_rfc_predicted))
```

```python
models = ['Logistic Regression', 'Support Vector Machine', 'Naive Bayes Classifier', 'Random Forest Classifier']
predictions = [y_lc_predicted, y_svc_predicted, y_nbc_predicted, y_rfc_predicted]
pred_probabilities = [y_lc_pred_proba, y_svc_pred_proba, y_nbc_pred_proba, y_rfc_pred_proba]

plot = 1

for model, prediction, pred_proba in zip(models, predictions, pred_probabilities):
    disp = ConfusionMatrixDisplay(confusion_matrix(y_test.ravel(), prediction))
    disp.plot(
        include_values=True,
        cmap='gray',
        colorbar=False
    )
    disp.ax_.set_title(f"{model} Confusion Matrix")

plt.figure(figsize=(30, 15))
plt.suptitle("ROC Curves")
plot_index = 1

for model, prediction, pred_proba in zip(models, predictions, pred_probabilities):
    fpr, tpr, thresholds = roc_curve(y_test, pred_proba[:, 1])
    auc_score = auc(fpr, tpr)
    plt.subplot(3, 2, plot_index)
    plt.plot(fpr, tpr, 'r', label='ROC curve')
    # pyplot.figure(figsize=(5, 5))
    plt.title(f'Roc Curve - {model} - [AUC - {auc_score}]', fontsize=14)
    plt.xlabel('FPR', fontsize=12)
    plt.ylabel('TPR', fontsize=12)
    plt.legend()
    plot_index += 1
plt.show()
```
